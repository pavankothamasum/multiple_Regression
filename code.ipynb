```python
import pandas as pd
import matplotlib.pyplot as plt 
from sklearn.model_selection import train_test_split 
```

```python
df=pd.DataFrame({
    'sqft':[2000,1600,2400,1800,3000,1500,2700,2200,1400,2500],
    'bedrooms':[3,3,4,3,4,2,4,3,2,4],
    'age_of_house':[10,15,5,8,12,20,4,6,18,3],
    'price(in$)':[450,300,520,400,600,250,550,475,280,500]
})
```

```python
df
```

```python
df.info()
```

```python
df.describe()
```

```python
x=df[['sqft','bedrooms','age_of_house']]
y=df[['price(in$)']]
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2,random_state=19)
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
```

```python
model.fit(xtrain,ytrain)
```

```python
model.predict(xtest)
from sklearn.metrics import mean_absolute_error 
print(mean_absolute_error(ytest,model.predict(xtest)))
```
